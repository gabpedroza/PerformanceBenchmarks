```python
import serial
arduino = serial.Serial(port='COM7',   baudrate=9600, timeout=.1)
x = []
y = []
cont = 0
while cont < 3:
    content = arduino.readline().rstrip()
    if(b", " in content and len(content) < 50):
        data = str(content.decode()).split(", ")
        x.append(int(data[0]))
        y.append(float(data[1]))
        print(f"{len(x)}: ({data[0]}, {data[1]})")
        cont += 1
```

```python
import matplotlib.pyplot as plt
plt.figure(0)
plt.bar([str(i) for i in x[::-1]], y[::-1])
plt.xlabel("Frequency (MHz)")
plt.ylabel("Time to complete (microseconds)")
plt.title("Time to traverse array")
plt.figure(1)
plt.bar([str(i) for i in x[::-1]], [i*a for i, a in zip(y[::-1], x[::-1])])
plt.xlabel("Frequency (MHz)")
plt.ylabel("Normalised time")
plt.title("Normalised time to complete")
plt.show()
```
