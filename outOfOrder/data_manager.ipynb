```python
toPlot = {}
with open("data.txt", "r") as file:
    for line in file:
        data = line.split(" ")
        data = [int(data[0]), int(data[1])]
        if data[0] not in toPlot.keys():
            toPlot[data[0]] = [data[1]]
        else:
            toPlot[data[0]].append(data[1])
for entry in toPlot.keys():
    avg = sum(toPlot[entry])/len(toPlot[entry])
    toPlot[entry] = avg

print(toPlot)
```

```python
import matplotlib.pyplot as plt
plt.bar(toPlot.keys(), toPlot.values())
plt.xlabel("Number of dependent sums")
plt.ylabel("Time (nanoseconds)")
plt.title("Time to traverse array with mul and adds")
plt.xticks([i for i in range(-1, 11)])
plt.show()
```
