```python
import serial
arduino = serial.Serial(port='COM7',   baudrate=9600, timeout=.1)
timeDualCore = 0
cont = 0
while cont < 1:
    content = arduino.readline().rstrip()
    if(b" @ " not in content and len(content) > 0):
        timeDualCore = float(str(content.decode()))
        cont += 1
```

```python
import serial
arduino = serial.Serial(port='COM7',   baudrate=9600, timeout=.1)
timeSingleCore = 0
cont = 0
while cont < 1:
    content = arduino.readline().rstrip()
    if(b" @ " not in content and len(content) > 0):
        timeSingleCore = float(str(content.decode()))
        cont += 1
```

```python
import matplotlib.pyplot as plt

plt.bar(["Single", "Dual"], [timeSingleCore, timeDualCore])
plt.title("Time to find largest integer")
plt.ylabel("Time (microseconds)")
plt.xlabel("Number of cores")
plt.show()
```
