```python
y1, y2 = [], []
count = 0
with open("results.txt", "r") as file:
    for line in file:
        data = line.split(" ")[1]
        if(count < 100):
            y1.append(int(data))
        else:
            y2.append(int(data))
        count += 1
```

```python
import matplotlib.pyplot as plt
plt.bar(["No SIMD", "SIMD"], [sum(y1)/len(y1), sum(y2)/len(y2)])
plt.ylabel("Time (nanoseconds)")
plt.title("Time to sum array")
```
