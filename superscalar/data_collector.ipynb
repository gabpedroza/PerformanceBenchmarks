```python
import matplotlib.pyplot as plt


x, y = [], []
with open("data.txt", "r") as file:
    for line in file:
        data = line.split(" ")
        #print(data)
        x.append(int(data[0]))
        y.append(int(data[1]))

def avg(i, lis):
    return sum([k for k in lis[i::7]])/len(lis[i::7])

plt.figure(0)
plt.bar([i for i in range(1, 8)], [avg(i, y) for i in range(0, 7)])
plt.xlabel("Number of buckets")
plt.ylabel("Time to sum (nanonseconds)")
plt.title("Time to sum array with $N$ buckets")
plt.show()
plt.figure(1)
plt.xlabel("Number of buckets")
plt.ylabel("Normalised time to sum (nanonseconds)")
plt.title("Normalised time to sum array with $N$ buckets")
plt.bar([i for i in range(1, 8)], [avg(i, y)*(i+1) for i in range(0, 7)])
plt.show()
```
